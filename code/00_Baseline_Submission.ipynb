### Imports

```python
import pandas as pd
from BaselineModel import BaselineModel
from score import *
```

### Load data

```python
df = pd.read_parquet("../data/train_series.parquet")
solution = pd.read_csv('../data/train_events.csv')
solution.dropna(subset=['step'], inplace=True)
```

### Baseline 

```python
baseline = BaselineModel(df)
submission = baseline.predict()
submission.head()
```

### Calculate score

```python
sample_val = solution[solution['series_id'].isin(submission.series_id.unique())]

sleep_tolerances = {
    'onset': [12, 36, 60, 90, 120, 150, 180, 240, 300, 360],  
    'wakeup': [12, 36, 60, 90, 120, 150, 180, 240, 300, 360]  
}

score = score(
        solution = sample_val,
        submission = submission,
        tolerances = sleep_tolerances,
        series_id_column_name = 'series_id',
        time_column_name = 'step',
        event_column_name = 'event',
        score_column_name = 'score',
        use_scoring_intervals =False,
)
score
```
